```python
#Q14
def string_to_dict(s):
    result_dict = {}
    for letter in s: # iterate through every character including spaces
        if letter != " ": # ignores spaces
          if letter in result_dict:
            result_dict[letter] += 1 # value+1 , in this case if letter exist with value 1 it will become 2
          else:
            result_dict[letter] = 1
    return result_dict        
            
s = input("Enter your string: ") 
response = string_to_dict(s)
print(response)  
```
