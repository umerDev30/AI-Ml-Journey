```python
#  reading and writing together
with open("jerry.txt","w+") as file:
    file.write("Hello jerry\n")
    print(file.read())
    file.write("how are you?")
    file.seek(0) # seek() moves the file cursor to a specific position in a file,The file cursor (or pointer) tells Python where to read or write next.
    content = file.read()
    print(content)
 
# file.seek(offset, whence)   
# | Parameter | Meaning                                                                                  |
# | --------- | ---------------------------------------------------------------------------------------- |
# | `offset`  | Number of bytes to move                                                                  |
# | `whence`  | Reference point: <br> 0 = beginning (default)<br>1 = current position<br>2 = end of file |

##############################################################################################################

# Why seek(0) is necessary

# After f.write(), the cursor is at the end of the file.

# f.read() will read from cursor to end.

# Without seek(0), f.read() returns empty string.

# with open("example.txt", "w+") as f:
#     f.write("Hello World\n")
#     print(f.read())  # ❌ prints nothing

# Mode	Meaning
# r+	Read & write, file must exist, doesn’t erase
# w+	Write & read, creates/overwrites file
# a+	Append & read, creates file if not exist, adds to end
```

```python
# Using seek() with tell()

# tell() → tells current position of cursor in the file

with open("jerry.txt", "r") as f:
    print(f.tell())   # 0 (start)
    f.read(10)
    print(f.tell())   # 10 (after reading 10 chars)
    f.seek(0)
    print(f.tell())   # 0 (back to start)
```

```python
def count_file_stats(filepath):
    with open(filepath,'r') as file :
      lines = file.readlines() # Grabs every line and puts it in a list
      num_of_lines = len(lines)
      num_of_words = sum(len(line.split()) for line in lines) # is called a generator expression.
      num_of_chars = sum(len(line) for line in lines) # here each line is a string and string is already iterable (like a list of characters)
    return num_of_lines, num_of_words, num_of_chars

s = count_file_stats('jerry.txt')
print(s)


# What a generator expression does:
# It produces numbers one by one

# It does NOT store them in a list

# sum() takes each number as it comes and adds it

# Step-by-step in slow motion

# Imagine lines has 3 lines:

# lines = [
#     "I like apples\n",
#     "You like bananas\n",
#     "Python is fun\n"
# ]


# The generator produces values like this:

# First line  → len(...) = 3
# Second line → len(...) = 3
# Third line  → len(...) = 3


# These numbers flow directly into sum():

# sum starts at 0
# 0 + 3 = 3
# 3 + 3 = 6
# 6 + 3 = 9


# ⚠️ At NO point is [3, 3, 3] stored anywhere

# 0 is the default starting value of the sum() and It keeps incrementing as each number comes from the iteration.

###You can even change the starting value:###

# sum([1, 2, 3], 10)


# Result:

# 16


# Because:

# start = 10
# 10 + 1 + 2 + 3 = 16


# Use readline() → when reading line by line manually

# Use readlines() → when you want all lines as a list

# Use read() → when you want everything as one string
```
