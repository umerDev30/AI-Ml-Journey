```python
# 1️⃣ Default Packages (Standard Library)

# Default packages are the packages that come pre-installed with Python.

# 📌 You do NOT need to install them using pip or uv.

# Examples
# import math
# import os
# import sys
# import datetime
# import random

# Key points

# Come with Python itself

# Written and maintained by Python core developers

# Always available once Python is installed

# Called Python Standard Library

# 🧠 Example:

# import math
# print(math.sqrt(16))

# 2️⃣ Open-Source Packages (Third-Party Packages)

# Open-source packages are not included by default.
# They are:

# Created by the community

# Publicly available

# Source code is open (you can read/modify it)

# 📌 You must install them manually.

# Examples
# pip install numpy
# uv pip install pandas

# import numpy as np
# import pandas as pd

# Popular open-source packages

# numpy → numerical computing

# pandas → data analysis

# matplotlib → plotting

# scikit-learn → ML

# tensorflow, torch → deep learning

# requests → HTTP requests

# fastapi, django → web development
```

```python
# Data Serialization with JSON
# Data serialization, such as converting Python objects (dict, list, etc.) into a JSON format,it is important when working with APIs or databases.

import json

data = {'name': 'umer','age':25}

json_string = json.dumps(data) # Python → JSON string
print(json_string)
print(type(json_string))

parse_json = json.loads(json_string) # JSON string → Python
print(parse_json)
print(type(parse_json))
```

```python
import csv

file_name = 'data.csv'
with open(file_name,'w',newline='') as file: # Disables Python’s automatic newline handling,Lets the CSV module control newlines
    write = csv.writer(file)
    write.writerows([['name','age'],['umer',24]])
    
with open(file_name,'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row) 
```

```python
from datetime import datetime, timedelta

now = datetime.now()
print(now)
yesterday = now - timedelta(days=1) # Take the current moment (now) and go back exactly 24 hours
print(yesterday)
```

```python
# Program Timing with Time Module The time module can be used to measure time and make the program sleep for a specified number of seconds.
# Print the current timestamp → wait 2 seconds → print the new timestamp
# measuring elapsed time and pausing execution.Elapsed time = how much time has passed between two moments/ there difference.
import time

print(time.time()) # it returns Number of seconds since 1 Jan 1970 (UTC), This reference point is called the Unix Epoch.
time.sleep(2)
print(time.time())
```

```python
import re

pattern = r'\d+'
test = "I have 124 strawberries."

match = re.search(pattern, test)
if match:
    print(match.group()) # match.group() → returns the actual matched substring
    
# pattern = r'\d+'

# \d → matches any digit (0-9)

# + → matches one or more of the preceding token

# So \d+ = one or more consecutive digits

# r'' → raw string, so \d is treated literally in regex    
```
