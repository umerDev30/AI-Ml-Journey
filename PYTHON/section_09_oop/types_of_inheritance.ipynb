# Types of Inheritance

1- Simple
2- Heirarichal
3- Multilevel
4- Multiple
5- Hybrid

```python
# Simple (Parent <-- Child)

class Grandfather:
     def __init__(self, name, age):
        self.name = name
        self.age = age

     def show_details(self):
        return f'Name: {self.name}, Age: {self.age}'

     def speak(self):
        return 'Grandfather speaks wisely.'
```

```python
class Father(Grandfather):
     def __init__(self, namee, age, occupation):
        super() .__init__(namee, age)
        self.occupation = occupation

     def show_occupation(self):
        return f'Occupation: {self.occupation}'

     def speak(self):
        return 'Father speaks carefully.'
```

```python
# Creating an object of the Father class
father_obj = Father('John', 50, 'Engineer')
print(father_obj.show_details())
print(father_obj.show_occupation())
print(father_obj.speak())
```

```python
## Heirarichal inheritance just involve more than 1 child. 
## Multi-level inheritance is like Grand parents <-- Parents <-- Child. child can get whatever grand parents have but through prents. 
```

```python
## Multiple

class Father:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show_details(self):
        return f'Name: {self.name}, Age: {self.age}'

    def speak(self):
        return 'Father speaks carefully.'

class Mother:
    def __init__(self, name, favorite_food):
        self.name = name
        self.favorite_food = favorite_food

    def show_favorite_food(self):
        return f'Favorite Food: {self. favorite_food}'

    def speak(self):
        return 'Mother speaks lovingly.'

class Child(Father, Mother): # order of inheritance matters
    def __init__(self, name, age, favorite_food, hobby):
        Father.__init__(self, name, age)
        Mother.__init__(self, name, favorite_food)
        self. hobby = hobby
        
    def show_hobby(self):
        return f'Hobby: {self.hobby}'

    # def speak(self):
    #     return 'Child speaks excitedly.'

```

```python
# Creating an object of the Child class
child_obj = Child('Daisy', 16, 'Pizza', 'Dancing')
print(child_obj.show_details())
print(child_obj.show_favorite_food())
print(child_obj.show_hobby())
print(child_obj.speak()) # father bez it is iherit first 
```

```python
# MRO method = the order in which Python looks for methods when you call a method on an object in multiple inheritance.

# Python follows a specific algorithm called C3 Linearization.

class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")

class C(A):
    def show(self):
        print("C")

class D(B, C):
    pass

d = D()
d.show()
```

```python
print(D.mro())
```

```python
class A:
    def __init__(self):
        self.value = 'A'

class B(A):
    def __init__(self):
        super().__init__()
        self.value = 'B'

class C(A):
    def __init__(self):
        super().__init__()
        self.value = 'C'

class D(B, C):
    def __init__(self):
        super().__init__()

obj = D()
print(obj.value)

# on D(B,C), so B constructor says super is next line to be executed can not go down to see value, so, C is next to me here  D(B,C) so visist C first
# ok i guess whatever channel it uses to go find will come through them as well no matter if value gets modified in between
# Conceptual analogy

# Think of super() as “go do the next class’s work first”

# After that work finishes, I (current class) continue my own work

# Anything you assign after super() will overwrite the values set by the next class
```
