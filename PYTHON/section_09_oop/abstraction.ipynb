```python
# Abstraction = show what an object can do, hide how it does it

# You expose only essential features and hide implementation details.

# Real-life example

# You use a remote control

# You press ON

# You don’t care how the circuit works


# Abstract classes define a contract:

# An abstract method MUST be overridden in a concrete (non-abstract) child class.
# An abstract method:
# Defines WHAT must exist
# Does NOT define HOW it works
# So the child class is responsible for the implementation.

# Abstract methods MUST be implemented by all concrete child classes

# Non-abstract methods are optional (you can use them as-is or override them)

# The user/developer uses what is exposed without worrying about how it’s implemented internally

# Non-abstract methods → use as-is OR override (optional)

# If a child class does NOT implement all abstract methods, then: The child class itself becomes abstract. You cannot create its object


### Abstract classes are useful when you have a group of related objects that must share common behavior (a contract), but each object implements it differently.
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    kingdom = "Animalia"        # class variable

    def __init__(self, name):   # constructor
        self.name = name

    def sleep(self):            # normal method
        print("Sleeping")

    @abstractmethod
    def sound(self):            # abstract method
        pass
```
