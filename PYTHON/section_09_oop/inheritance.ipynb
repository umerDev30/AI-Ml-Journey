```python
class Human:
    def __init__(self,name,age):
        self.name = name
        self._age = age
    
    def canSwim(self):
        print("Humans can swim")
    def _run(self):
        print("Running")   
    def canCook(self):
        return True
     
```

```python
class Male(Human): 
    def canSwim(self):
        print("Males can swim")
    def breadwinner(self):
        return True
    def age(self):
        return f"Age of Male is{self._age}"
```

```python
m = Male("Ali",26)
```

```python
m.canSwim() # when we have same function defined in parent and sub class, and we make obj of subclass then call the function which exist in both classes; The function of sub class will overidethe parent function.Same method name → child overrides parent → child wins
# That concept is called method overriding.   
```

```python
# Constructor Overriding

class Parent:
    def __init__(self):
        print("Parent init")

class Child(Parent):
    def __init__(self): # Child constructor overrides parent constructor unless super().__init__() is called.
        print("Child init")

c = Child()

```

```python
class Child(Parent):
    def __init__(self):
        super().__init__() # super keyword is used to access methods of parent class from our child class. ONLY METHODS/CONSTRUCTOR.cannot be used outside class
        print("Child init")

c = Child()
```

```python
class User:
      def __init__(self,name,Id):
        self.name = name
        self.Id = Id
      def login(self):
        print("P logged in")
      def logout(self):
        print("P logged out")
   
```

```python
class Student(User):
       def __init__(self,marks,roll_number,name,Id): 
            super().__init__(name,Id) # self is implicit in method calls — never pass it manually
            self.marks = marks
            self.roll_number = roll_number
            
       def login(self): # this function will be called; do whatever its need to do and then this super.login()# parent function will do its job.
           print(("Student Logging in"))
           super().login() 
```

```python
s = Student(68,311,"Umer","100")
print(s.name)
s.login()
```
