## why this error?

Create a new object of type Student.
Call the constructor of that class, passing the new object itself as the first argument.

Python tries to call your method like this:

Student._init_(<new Student object>)


Python passed one argument (the new object) automatically.

Your _init_ definition expects zero arguments, because you didn’t write self.

Hence the error:

TypeError: Student.__init__() takes 0 positional arguments but 1 was given


✅ Key: You didn’t pass anything, but Python always passes the object as the first argument for instance methods.

3️⃣ Why it matters

self is the handle to the object inside the method.

Without it, Python doesn’t know where to store attributes or how to link the object to the method.

The error occurs because the number of parameters doesn’t match the number of arguments Python sends.

```python
class Student:
    def __init__(self,namee,age,roll_number): # here we are taking value from paramter and saying this obj has this property(i.e:self.nameeee) and that property is equal to this parameter which will be taken while obj creation
        self.nameeee = namee # we keep these same but inorder to distinguish we named them different
        self.age = age
        self.roll_number = roll_number
        print("I will get called automatically when you create an object")   
        
    def study(self):
        print("I am "+self.name+" and I study at UOK.")
    def play(self):
        print("I play")            
```
