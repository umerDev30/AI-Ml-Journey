1️⃣ Procedural Programming
🔹 What it is

Procedural programming is a style where you write step-by-step instructions (procedures/functions) that tell the computer what to do and in what order.

Think of it like a recipe:

Take ingredients

Mix them

Cook

Serve

The program runs top to bottom, following commands in sequence.

🔹 Key ideas

Program is divided into procedures/functions

Uses variables that can change (state)

Focus is on HOW to do something

Order of execution matters

````python
# Example

def add(a, b):
    return a + b

def average(numbers):
    total = 0
    for n in numbers:
        total += n
    return total / len(numbers)

nums = [1, 2, 3, 4]
result = average(nums)
print(result)



# 🔹 Pros

# ✅ Easy to understand
# ✅ Good for small & medium programs
# ✅ Fast to write

# 🔹 Cons

# ❌ Hard to manage in very large programs
# ❌ Changing one part can affect many others
````

2️⃣ Functional Programming
🔹 What it is

Functional programming is a style where you build programs using pure functions that do not change data, but instead return new data.

Think of it like math:

f(x) = x + 2


Same input → Same output
No side effects.

🔹 Key ideas

Everything is based on functions

No changing variables (immutability)

Focus is on WHAT to do, not HOW

Avoids side effects (no unexpected changes)


````python
# Example 1

def average(numbers):
    return sum(numbers) / len(numbers)

nums = [1, 2, 3, 4]
print(average(nums))

#  Example 2

numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x*x, numbers))
print(squared)

# 🔹 Real-life analogy

# 🧠 Vending machine

# You put money + choose item

# Machine gives item

# Machine state doesn’t care about previous user

# Same input → Same output.

# 🔹 Pros

# ✅ Fewer bugs
# ✅ Easier to test
# ✅ Safer for parallel & AI systems
# ✅ Cleaner logic

# 🔹 Cons

# ❌ Harder for beginners
# ❌ Can feel confusing at first
# ❌ Sometimes slower or more abstract
````

🔹 Definition (very important)

A pure function is a function that:

1️⃣ Always gives the same output for the same input
2️⃣ Does NOT change anything outside itself

🔹 Example of a PURE function
def square(x):
    return x * x


Input: 4

Output: 16

Every time → same result

Does not touch any external variable

✅ This is pure



````python
# 🔹 Example of an IMPURE function
count = 0

def increment():
    global count
    count += 1
    return count


# ❌ Output depends on previous calls
# ❌ Changes external variable (count)
# ❌ NOT pure
````

String object literal

Numeric object literal

Boolean object literal

None literal
x = None      # None literal → Python creates a NoneType object

✅ So basically, an object literal is a literal value that Python immediately turns into an object.

*1️⃣ What is Object-Oriented Programming (OOP)?*  
OOP is a programming paradigm that uses "objects" — instances of classes — to model real-world entities.  
It organizes code by bundling data and behavior (functions) together.  
Popular OOP languages: Java, Python, C++, C#.

*2️⃣ What are the main principles of OOP?*  
There are four core pillars:  
• *Encapsulation* – Bundling data with methods  
• *Abstraction* – Hiding internal details  
• *Inheritance* – Reusing code from parent classes  
• *Polymorphism* – Using one interface for many forms  

*3️⃣ Define Class and Object.*  
• *Class*: A blueprint or template for creating objects (e.g., `Car`, `Person`)  
• *Object*: A real instance of a class with specific data and behavior  
Example:  
```python
class Car:  
    def _init_(self, brand):  
        self.brand = brand  

my_car = Car("Toyota")  # Object
```

*4️⃣ What is the difference between Class and Structure?*  
In C++:  
• Class: Members are private by default  
• Structure: Members are public by default  
• Classes support full OOP features; structs are more lightweight  
In modern languages like C#, both are similar but structs are *value types*, classes are *reference types*.

*5️⃣ What is Encapsulation?*
Encapsulation means hiding the internal state and requiring all interaction through methods.  
It improves security and control.  
Example:
```python
class Account:  
    def _init_(self):  
        self.__balance = 0  # private

    def deposit(self, amount):  
        if amount > 0:  
            self.__balance += amount
```

*6️⃣ What is Abstraction?*  
Abstraction means exposing only relevant information while hiding complexity.  
Focuses on *what* an object does, not *how*.  
Example:  
You can use `print()` without knowing how it sends text to the screen.

*7️⃣ What is Inheritance?*  
Inheritance allows a class (child) to acquire properties and behaviors of another class (parent).  
Reduces redundancy.  
Example:
```python
class Animal:  
    def speak(self):  
        print("Animal sound")

class Dog(Animal):  
    pass

Dog().speak()  # Inherits from Animal
```

*8️⃣ What is Polymorphism?*  
Polymorphism means "many forms".  
Same function name or operator can behave differently based on context.  
Example: `+` adds numbers or concatenates strings.

*9️⃣ Types of Polymorphism?*  
• *Compile-time* (Static): Method overloading, operator overloading  
• *Run-time* (Dynamic): Method overriding via inheritance  

*🔟 What is Method Overloading?*
Defining multiple methods with the same name but different parameters (not supported natively in Python but possible via default arguments).  
In Java/C++:  
```java
void greet()  
void greet(String name)
```
