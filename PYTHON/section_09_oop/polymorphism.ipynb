```python
# 1️⃣ What is Polymorphism?

# Polymorphism = “many forms”

# In Python OOP:

# Same method name can behave differently depending on:

# The object calling it (runtime polymorphism)

# The parameters passed (compile-time/overloading, but Python doesn’t support strict overloading)

# Polymorphism = flexible behavior using same interface
```

```python
# 2️⃣ Types of Polymorphism
# A. Compile-time (Method Overloading)

# Python does NOT support true overloading like Java/C++

# You can achieve similar behavior using:

# Default arguments

# *args, **kwargs

# class Calculator:
#     def add(self, a, b=0):
#         return a + b

# c = Calculator()
# print(c.add(5))      # 5
# print(c.add(5, 10))  # 15
```

```python
class User:
    def __init__(self, name, mobileNumber, address=None):
        if address is None:
            self.C1(name, mobileNumber)
        else:
            self.C2(name, mobileNumber, address)

    def C1(self, name, mobileNumber):
        self.name = name
        self.mobileNumber = mobileNumber

    def C2(self, name, mobileNumber, address):
        self.name = name
        self.mobileNumber = mobileNumber
        self.address = address

# Create object
u1 = User("Mayank", "9999999", "Delhi")
print(u1.name)         # Mayank
print(u1.mobileNumber) # 9999999
print(u1.address)      # Delhi
```

```python
# B. Runtime (Method Overriding)

# Child class overrides parent method

# Method called depends on object type, not reference type

class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def greet(self):
        print("Hello from Child")

p = Parent()
c = Child()

p.greet()  # Hello from Parent
c.greet()  # Hello from Child
```

```python
# C. Polymorphism with different classes

# Different classes can have methods with same name

# You can call them interchangeably without worrying about class type

class Dog:
    def speak(self):
        print("Woof")

class Cat:
    def speak(self):
        print("Meow")

def make_speak(animal):
    animal.speak()  # Works with any object having speak()

d = Dog()
c = Cat()

make_speak(d)  # Woof
make_speak(c)  # Meow


# ✅ This is polymorphism in action — same interface (speak) behaves differently depending on object type.
```

```python
# 1️⃣ What is Operator Overloading?

# **Operator overloading = giving special meaning to operators (+, -, *, etc.) for your custom classes.

# Normally + works on numbers or strings

# You can make + work on your objects by defining a special method

# 2️⃣ Special methods (dunder methods)
# Operator	Special Method
# +	__add__(self, other)
# -	__sub__(self, other)
# *	__mul__(self, other)
# /	__truediv__(self, other)
# ==	__eq__(self, other)
# <	__lt__(self, other)
# <=	__le__(self, other)
# str()	__str__(self)
# repr()	__repr__(self)

# Basically, almost any operator can be “overloaded” for your class.

# 3️⃣ Example with a ComplexNumber class
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        return ComplexNumber(self.real + other.real, self.imag + other.imag)

    def __str__(self):
        return f"{self.real}+{self.imag}i"

c1 = ComplexNumber(2, 3)
c2 = ComplexNumber(1, 4)

print(c1 + c2)  # 3+7i


# ✅ Here, + is overloaded to work with ComplexNumber objects.

# 4️⃣ Key points

# Overloaded operators work only if you define the corresponding dunder method

# Both operands can be custom objects

# You can overload arithmetic, comparison, logical, and even indexing operators

# 5️⃣ Another example — comparison
class Student:
    def __init__(self, marks):
        self.marks = marks

    def __gt__(self, other):
        return self.marks > other.marks

s1 = Student(85)
s2 = Student(90)

print(s1 > s2)  # False


# ✅ Now > works for Student objects based on marks.

# 6️⃣ One-line rule 🧠

# Operator overloading = define a dunder method in your class → Python uses it when that operator is applied to your object
```
