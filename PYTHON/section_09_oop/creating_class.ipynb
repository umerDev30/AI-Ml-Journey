```python
class ComplexNumber:
    def __init__(self,real = 0.0 ,imag = 0.0):
        self.imag = imag 
        self.real = real 
        
    def conjugate(self):
        return ComplexNumber(self.real,-self.imag) # New complexNumber returned
        
    def __str__(self): # is a special method that tells Python,How should this object(self) look when I print it.You can store it, concatenate it, put it in a list, or use it anywhere a string is needed.
        if (self.real == 0): #__str__() does NOT store anything and does NOT change your object.It only returns a string temporarily when Python wants to display the object.
          s = f"{self.imag}i"
        elif (self.imag < 0):
          s = f"{self.real}{self.imag}i" 
        else:    
          s = f"{self.real}+{self.imag}i"    
        return s
      
    def __add__(self,other): # self will take left operand(obj ref) and other will take right operand(another obj ref).
      finalReal = 0 # just to initialize variables, however not necessary here
      finalImag = 0
      finalReal = self.real + other.real
      finalImag = self.imag + other.imag
      resultComplexNumber = ComplexNumber(finalReal,finalImag)
      return resultComplexNumber
    
    def __sub__(self,other): 
      finalReal = self.real - other.real
      finalImag = self.imag - other.imag
      resultComplexNumber = ComplexNumber(finalReal,finalImag)
      return resultComplexNumber
    
    def __mul__(self,other): # (a+bi)(c+di)=(ac-bd)+(ad+bc)i
      finalReal = self.real * other.real - self.imag * other.imag # ac - bd 
      finalImag = self.real * other.imag + self.imag * other.real # ad + bc
      resultComplexNumber = ComplexNumber(finalReal,finalImag)
      return resultComplexNumber
# -7 + 22i
# (real = -7, imag = 22)
        
    # def __truediv__(self,other): 
    #   finalReal = (self.real * other.real + self.imag * other.imag) / ((other.real)**2+(other.imag)**2) # ac + bd/c2 + d2 
    #   finalImag = (self.imag * other.real - self.real * other.imag) / ((other.real)**2+(other.imag)**2) # bc -ad/c2 + d2
    #   resultComplexNumber = ComplexNumber(finalReal,finalImag)
    #   return resultComplexNumber 
     
    def __truediv__(self,other): 
      denominator = (other.real)**2+(other.imag)**2
      resultComplexNumber = self * ComplexNumber(other.real/denominator,-other.imag/denominator) #Python checks if ComplexNumber has a __mul__ method.It calls your __mul__ method, passing the self as self and other complexNumber as other
      return resultComplexNumber 
       
    ## TRICK FOR DIVISION
    #Numerator: (a+bi) * (c-di)
    #Denominator: c^2 + d^2

    #3️⃣ Divide first to simplify
    #Instead of dividing after multiplication, you can divide the conjugate by the denominator upfront:
    # c-di/dem= c/dem - d/dem i
    # self is (a + bi)
    #ComplexNumber(other.real/den, -other.imag/den) is (c - di) / (c^2 + d^2)
    # * uses your __mul__ function to multiply the two complex numbers.
    
    def __eq__(self, other):
       return (self.real == other.real) and (self.imag == other.imag)
    
# __str__() → gives you a string representation of the object, so you can store it, reuse it, or print it anywhere

## when c = complex()
# print(c)
# real + imag*j
# Here, 0.0 + 0.0j → simplified → 0j
```

1️⃣ Python order of operations (simplified)

Python follows PEMDAS (Parentheses → Exponents → Multiplication/Division → Addition/Subtraction).

Parentheses () first ✅

Exponentiation ** ✅

Multiplication * and Division /, left to right ✅

Addition + and Subtraction -, left to right ✅
