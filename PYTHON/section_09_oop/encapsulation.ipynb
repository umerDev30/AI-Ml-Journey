```python
# Encapsulation ==> Hiding the internal details of a class and controlling access to its data and methods.

# Why we use it

# Protect data from accidental changes

# Control access (read-only, write-only, restricted)

# Maintain consistency (you can validate values before changing them)

# Improve maintainability (hide internal complexity)

# #Access modifiers ==> Access modifiers are keywords or conventions that control the visibility of class members (variables and methods) from outside the class.
```
