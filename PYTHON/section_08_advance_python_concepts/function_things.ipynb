```python
## Decorators

def decorator(func):
    def wrapper(*args,**kwargs):
        args_value = ", ".join(map(str,args)) if args else "0" # Apply str() to each element in args. That’s a loop — just hidden. Join()= glue args together using ", "
        kwargs_value = ", ".join(f"{k}:{v}" for k,v in kwargs.items()) if kwargs else "0"
        print(f"Function name is {func.__name__} "
            f"and args are {args_value}, kwargs are {kwargs_value}.")
        return  func(*args,**kwargs)
    return wrapper
@decorator
def greetings(name,greet='Hello'):
    return f"{name}, {greet}"

# *args → tuple (never None),**kwargs → dict (never None).Decorators replace the function with wrapper
# if args = “If args is NOT empty”

#if not args = “If args IS empty”

## if args: means “arguments exist”, and if not args: means “no arguments exist”.
```

```python
greetings("Umer")
```

```python
#Ex 3
import time


def decorator(func):
    cache_value = {}
    def wrapper(*args):
        if args in cache_value:
            return cache_value[args]
        result = func(*args)
        cache_value[args] = result # The dictionary keys are the args tuple, created automatically when the function is called.
        return result
    return wrapper

@decorator
def long_running_func(a,b):
    time.sleep(5)
    return a + b

# Why can args be a dictionary key?

# Because:

# Tuples are immutable

# Immutable objects are hashable

# Hashable objects can be dictionary keys

# cache_value[(2, 3)] = 5

# Cache now:
# { (2, 3): 5 }
```

```python
# Bonus (pro tip)

# Python already has this built-in:

# from functools import lru_cache

# @lru_cache
# def long_running_func(a, b):
#     ...


# Same concept, production-ready.
```

```python
print(long_running_func(2,8))
print(long_running_func(5,3))
print(long_running_func(2,8))
```
