```python
# A generator is a function that pauses, remembers its state, and resumes later.
# It is a nice way to create iterators
```

```python
def gen(limit):
    current = 1
    while current <= limit:
        yield current
        current +=1
        
# Same logic

# current → state

# yield → return + pause

# loop end → StopIteration automatically

## Difference b/w return and yeild

# return

# Ends the function forever

# State is lost

# yield

# Returns a value

# Freezes the function

# Saves:

# local variables

# current line number

# Resumes from that exact spot
```

```python
g = gen(5) # Nothing runs yet.
```

```python
print(next(g))
```

```python
print(next(g))
```

```python
# another way

def gen(limit):
    for i in range(limit):
        yield i
```

```python
g = gen(5)
```

```python
print(next(g))
```

```python
print(next(g))
```

```python
g = (i**2 for i in range(3))
```

```python
print(next(g))
```

```python
#ex 
def my_gen():
    yield 1
    yield 2
    yield 3
```

```python
g = my_gen()
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```
