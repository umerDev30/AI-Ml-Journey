```python
#Q1
def temperature_convert(magnitude,unit):
    unit = unit.upper()
    if unit == "C":
        return (magnitude * 9/5) + 32
    elif unit == "F":
        return (magnitude -32) * 5/9
    else:
        return None
```

```python
print(temperature_convert(32,"c"))
```

```python
#Q2
def password_checker(password):
    positive_response = "Stronge Password"
    negative_response = "Weak Password"
    if len(password) < 8 :
        return negative_response
    if not any(char.upper() for char in password):
        return negative_response
    if not any(char.lower() for char in password):
        return negative_response  
    if not any(char.isdigit() for char in password):
        return negative_response
    special_char = "!@#$%^"
    if not any(char in special_char for char in password):
        return negative_response
    return positive_response
    
```

```python
p = password_checker("AC!1234vb")
print(p)
```

```python
#Q3
def billing(cart):
    total_cost = 0
    for item in cart:
        total_cost = total_cost + item['price']*item['quantity']
    return total_cost    
```

```python
cart = [
    {'name': 'Apple', 'price': 0.5, 'quantity': 4},
    {'name': 'Banana', 'price': 0.3, 'quantity': 6},
    {'name': 'Orange', 'price': 0.7, 'quantity': 3}
]
bill = billing(cart)
print(f"{bill:.2f} $")
```

```python
#Q4
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]

print(is_palindrome("madam"))
print(is_palindrome("umer"))
```

```python
#Q5
def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)
    
print(factorial(5))    
```

```python
#Q6
def count_word_frequency(file_path):
    word_count = {}  # 1. Create an empty dictionary to store word counts
    with open(file_path, 'r') as file:  # 2. Open the file in read mode
        for line in file:  # 3. Loop through each line in the file
            words = line.split()  # 4. Split the line into list of individual words
            for word in words:  # 5. Loop through each word
                word = word.lower().strip('.,:;!?/\\')  # 6. Make it lowercase and remove punctuation
                # 7. Update the count of the word in the dictionary
                word_count[word] = word_count.get(word, 0) + 1
    return word_count  # 8. Return the dictionary containing all word counts
 
```

```python
file_path = 'city.txt'
word_frequency = count_word_frequency(file_path)
print(word_frequency)
```

```python
#Q7

import re
def email_validate(email):
    pattern = r'^[A-Za-z0-9+_.-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
    return re.match(pattern,email) is not None
```

```python
print(email_validate('umer@gmail.com'))
```
