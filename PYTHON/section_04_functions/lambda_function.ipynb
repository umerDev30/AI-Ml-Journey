A lambda function in Python is a small anonymous function — meaning it doesn’t need a name. It’s often used for simple operations you want to do inline.
Lambda functions can have any number of arguments, but only one expression. They are commonly used for short operations or as arguments to higher-order functions.

Syntax:
lambda arguments: expression


arguments → input parameters (like normal function)

expression → the computation or return value (must be a single expression)

Returns the result automatically; no return keyword needed


```python
a = lambda z,y : y+z
print(a(2,4))
```

```python
even = lambda n : n % 2 == 0 
print(even(7))
```

The map function applies a given function to all items in an input list or any other iterable and returns a map object, which is an iterator. This is particularly useful for transforming data in a list comprehensively.

map(function, iterable) does not immediately compute all results.

Instead, it returns a map object, which is an iterator.

Values are calculated one by one as you loop over them or convert to a list.

```python
numbers = [1,2,3,4,5]
squared = map(lambda x : x**2,numbers) # squared is a map object, which is an iterator, not a list of numbers.
print(list(squared))

# or we can iterate
for i in squared:
    print(i)
```

```python
numbers1 = [1,2,3,4,5]
# or we can iterate
squared1 = map(lambda x : x**2,numbers1)
for i in squared1:
    print(i)
```

```python
def square(x):
    return x * x
list(map(square,[1,2,3]))
```

```python
list1 = [1,2,3,4]
list2 = [5,6,7,8]
add = map(lambda x,y : x+y , list1,list2)
print(list(add))
```

```python
str_numbers = ["1","2","3"]          # List of strings
int_numbers = map(int, str_numbers)  # map object (lazy)
print(list(int_numbers))              # Convert to list to see results

# Use parentheses () with int when you want to execute the function now.
# Don’t use parentheses when you want to pass the function itself as a parameter.
```

```python
fruits = ["apple","guava","orange"]
upper_case = list(map(str.upper,fruits))
print(upper_case)

# we will not put any function with () inside map
```

```python
details = [{'name': 'umer','age':25},{'name': 'jack','age':15}]
get_name = list(map(lambda x: x['name'], details))
print(get_name)

# map(lambda x: x['name'], details)
# lambda x: x['name'] is an anonymous function that takes a dictionary x and returns the value associated with the key 'name'.
# map() applies this lambda to each dictionary in the list.
# First dictionary: {'name': 'umer', 'age': 25} → 'umer'
# Second dictionary: {'name': 'jack', 'age': 15} → 'jack'
```

## Key Takeaways
The map function applies a specified function to all items in an iterable and returns a map object.
map can be used with regular functions, lambda functions, and multiple iterables for versatile data processing.
Built-in functions like int and string methods can be applied to iterables using map.
The map function enhances code efficiency and readability by eliminating the need for explicit loops.


## ✅ What is filter()?

filter() is used to select items from a list (or any iterable) based on a condition.

It keeps only the items for which the function returns True.

✅ Syntax
filter(function, iterable)


function → must return True or False

iterable → list, tuple, dictionary, etc.

filter() is lazy, so you usually convert the result into a list:

list(filter(...))


```python
num_list = [1,2,3,4,5]
filtering = list(filter(lambda x: x % 2 != 0 , num_list))
print(filtering)
```

```python
fruit_list = ['apple','banana','orange']
filtering_fruits = list(filter(lambda x : x.startswith('a'), fruit_list))
print(filtering_fruits)
```

```python
people = [{'name':'umer','age':25},{'name':"Albert",'age':19}]
adult = list(filter(lambda x : x['age'] > 18, people))
print(adult)
```

🎯 KEY DIFFERENCE to remember

map() → transforms each item

filter() → removes items based on a condition

The filter function in Python constructs an iterator from elements of an iterable for which a function returns true.
Filter is commonly used to filter out items from a list or other iterable based on a condition, supporting both named and lambda functions.
Multiple conditions can be applied within a filter using lambda expressions and logical operators.
The filter function can be used not only with lists but also with dictionaries and other iterables for data cleaning and manipulation.
