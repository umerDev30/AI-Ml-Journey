## Functions are block of code that performs a specific task.
# help organize code
# make code reusable
# improve readability

# 1️⃣ Actions (side effects) — no return needed

These are things a function does rather than gives back.

Examples:

printing something

updating a global variable

modifying a list/dictionary passed in

writing to a file

showing output on screen

✨ These actions already affect the outside world, so you don’t need to return anything.
i.e:
    def greet():
        print("Hello!")  # action

# 2️⃣ Values (results) — return IS needed

If you want the function to provide a value that you can store or use later, then you need return.
i.e:
def add(a, b):
    return a + b  # gives back a value        

## Here’s the key:

A function with no return ALWAYS returns None internally.

But you don’t see that None unless you print the function call itself.

🔍 Example to make it obvious
Case 1: Function prints but doesn’t return → works fine
def hello():
    print("Hi!")


Calling:

hello()


Output:

Hi!


✔ No None shown
✔ No problem
✔ Printing is fine

Case 2: You print the function call → NOW you see None
print(hello())


Output:

Hi!
None


First line: "Hi!" from inside the function

Second line: None because the function didn't return anything

🎯 Easy rule to remember
Using print() inside the function is perfectly fine.
Just don’t print( function_call ) unless the function actually returns something.

```python
# function with default parameters
def greet(name = "Whatever"):
    print(f"Hello {name}")
    
greet()    
```

## ⭐ Variable-Length Arguments (VLAs) in Python
1️⃣ *args → many positional arguments

Use when you don’t know how many values will be passed.

Example:
def add_all(*args):
    return sum(args)

print(add_all(1, 2, 3, 4))


➡ *args collects all extra positional arguments into a tuple.Python creates a tuple and puts all passed values inside it; internallly args = (1, 2, 3)

Easy memory tip:

*args → “star args” → unlimited numbers

2️⃣ **kwargs → many keyword arguments

Use when you don’t know how many key=value pairs will be passed.

Example:
def show_info(**kwargs):
    print(kwargs)


Calling:

show_info(name="Alex", age=20)


➡ **kwargs collects all extra keyword arguments into a dictionary.Python creates a dict and stores each key=value pair inside it.

## ⭐ *args — what values can it have?
ANY Python data type. Literally anything.

Because *args becomes a tuple, and tuples can store any mix of data types!

Examples of values *args may contain:

integers → 10

floats → 3.14

strings → "hello"

booleans → True

lists → [1, 2]

tuples → (3, 4)

dictionaries → {"a": 1}

custom objects → MyClass()

functions → len, print

even mixtures → (1, "hi", [2, 3], 4.5)

👉 Anything passed as positional arguments gets collected here.

⭐ **kwargs — what values can it have?
**Keys must be strings (technically identifiers),

but values can be ANY Python data type.**

Because **kwargs becomes a dictionary.


```python
def print_detail(*args, **kwargs):
    print("Positional arguments:")
    for val in args:
        print(val)
    print("Keyword arguments:")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_detail(1, 2, 3, 4, 5, name="Krish", age=32)
```

```python
def numbers(*args):
    return sum(args)

print(numbers(1,2,3))
```
