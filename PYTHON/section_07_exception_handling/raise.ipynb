```python
# What raise REALLY means

# raise = deliberately throw an exception

# Not because Python failed —
# but because YOU decided something is wrong.

# Think:

# “This situation is invalid. Stop normal execution.”

# Real-life analogy 🚨
# Airport security ✈️

# Scanner detects a prohibited item

# System raises an alarm

# Everything stops immediately

# The scanner didn’t “crash” —
# it intentionally raised an alert.

# Basic syntax
# raise ValueError("Invalid input")


# Python immediately:

# Stops execution

# Looks for a matching except

# If none → program crashes
```

```python
# Example 

age = int(input("Enter age: "))
if age < 0:
    raise ValueError("Age cannot be negative")
```

```python
# Example

try:
    amount = int(input("Enter amount"))

    if amount <= 0:
        raise ValueError("Amount must be positive")

except ValueError as e:
    print("Validation error:", e)

else:
    print("Transaction allowed")

finally:
    print("Session closed")

```

```python
# Example 
try:
   age = int(input("please enter age"))
   if(age<18):
        raise Exception("Please enter a age which is greater than 18") # this will be raised 
except TypeError as e:
    print(e)
except ValueError as e:
    print(e)
except Exception as e: # here it will be catched and printed
    print(e)
```

```python
# Example

num = 10
try:
    den = int(input("Enter den"))
    if(den == 0):
      raise ZeroDivisionError("Please enter a number which is not 0")
    print(num/den)
except ZeroDivisionError as e:
    print(e)
```
