# Handle specific errors (subclasses) first, and use Exception only as a last-resort safety net.

```python
# Think of Exception as “Something went wrong”
# And specific errors as “WHAT exactly went wrong”


## The REAL reason we separate exceptions
# Different problems need different actions

# Not different messages — different behavior.

try:
    user_age = int(input("Enter age: "))
    result = 100 / user_age

except ValueError:
    # USER mistake → ask again
    print("Please enter a valid number")

except ZeroDivisionError:
    # LOGIC issue → block execution
    print("Age cannot be zero")

except Exception:
    # SYSTEM issue → crash safely
    print("System error. Contact support")
    
# If you catch everything with Exception…

# 👉 How do you:

# Re-prompt input?

# Fix logic?

# Log system bugs?

# You can’t, because you don’t know the cause.   

## Exception is NOT for handling logic
# Exception is for damage control

# Think of it as:

# "I don’t know what happened, but don’t crash the system" 

```

```python
# 1️⃣ Real-life analogy (VERY intuitive)
# Scenario: ATM withdrawal 💳
# Step-by-step thinking:

# TRY
# 👉 You insert your card and enter PIN
# (“Let me try this risky operation”)

# EXCEPT
# 👉 If PIN is wrong → show error and stop
# 👉 If card is blocked → show error and stop

# ELSE
# 👉 PIN is correct → proceed with withdrawal

# Now inside ELSE, you still have risks:

# Bank server might be down

# Cash dispenser might jam

# So you need another try–catch inside ELSE.
```

```python
# 2️⃣ Mapping to Python (clean & realistic)

try:
    # Step 1: authentication (first risky step)
    pin = int(input("Enter ATM PIN: "))
    if pin != 1234:
        raise ValueError("Invalid PIN")

except ValueError as e:
    print("Authentication failed:", e)

else:
    # Step 2: transaction (new risky step)
    try:
        amount = int(input("Enter amount to withdraw: "))

        if amount <= 0:
            raise ValueError("Invalid amount")

        print("Processing transaction...")
        # simulate cash dispenser
        if amount > 5000:
            raise RuntimeError("ATM cash dispenser error")

        print("Cash dispensed successfully 💸")

    except ValueError as e:
        print("Transaction error:", e)

    except Exception as e:
        print("ATM system error:", e)
        
        
# ✅ Good design (nested tries = logical steps)

# Outer try → Authentication stage

# Outer else → Only runs if auth succeeded

# Inner try → Transaction stage

# Each stage:

# Has its own risks

# Has its own recovery logic

# This is exactly how backend systems are written.        
```

```python
# 1️⃣ What finally REALLY means (not the textbook line)

# finally runs no matter what happens
# — success, handled error, unhandled error, even return.

# Think of it as:

# “Before leaving this block, clean up.”

# 2️⃣ Real-life analogy (ATM again 💳)
# ATM flow:

# Insert card

# Enter PIN

# Withdraw cash

# Eject card

# 👉 No matter what happens:

# Wrong PIN ❌

# Server down ❌

# Cash dispensed ✅

# The card must be ejected.

# That’s finally.
```

```python
try:
    print("Insert card")
    pin = int(input("Enter ATM PIN: "))

    if pin != 1234:
        raise ValueError("Invalid PIN")

except ValueError as e:
    print("Authentication failed:", e)

else:
    # Runs only if PIN is correct
    try:
        amount = int(input("Enter withdrawal amount: "))

        if amount <= 0:
            raise ValueError("Invalid amount")

        print("Processing transaction...")
        if amount > 5000:
            raise RuntimeError("ATM dispenser error")

        print("Cash dispensed successfully 💸")

    except ValueError as e:
        print("Transaction error:", e)

    except Exception as e:
        print("ATM system error:", e)

    finally:
        # Inner cleanup
        print("Transaction session closed")

finally:
    # ALWAYS runs
    print("Ejecting card 💳")
```

```python
# 5️⃣ Why finally exists (core reasons)
# 1️⃣ Resource cleanup 🧹

# Close files

# Release DB connections

# Close network sockets

# Free locks

try:
    f = open("data.txt")
    process(f) # type: ignore
finally:
    f.close()


# Even if process() crashes → file is closed.

# 2️⃣ System stability

# In backend systems:

# Memory leaks

# Open connections

# Locked resources

# ❌ Without finally → system degrades over time
# ✅ With finally → predictable behavior

# 3️⃣ Separation of concerns
# Block	Purpose
# try	Risky operation
# except	Recovery
# else	Success logic
# finally	Cleanup

# This makes code readable and professional.

# 6️⃣ When to use finally (clear rules)
# ✅ Use finally when:

# You acquire something that must be released

# You need guaranteed execution

# You manage external resources

# ❌ Don’t use finally for:

# Normal logic

# Business rules

# Return values (unless intentional)

# 7️⃣ Important advanced behavior (interview favorite)
# Example

def ErrHandle():
    try:
       print("I am here, opening all the files")
       return "Success, got all files"
    except ZeroDivisionError as e:
       print(e)
    finally:
       print("Closing all the files")


# Output:

# I am here, opening all the files
# Closing all the files
# Success, got all files

# 👉 finally runs even after return

# ⚠️ If finally has its own return, it overrides the original return
# (which is usually a bad idea).

# 8️⃣ Senior-level golden rule 🔒

# If you open it, lock it, or start it — close it in finally.
```

```python
# What does “cleanup” ACTUALLY mean?

# Cleanup = undoing or safely finishing anything you started,
# so the system is left in a stable state.

# It is not business logic.
# It is not success logic.
# It is not error handling.

# It’s housekeeping.
```
