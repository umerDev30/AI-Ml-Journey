```python
class WrongEmailError(Exception):
    pass
```

```python
class Person:
  def __init__(self, salary, email):
    if(salary>10000000000):
      raise ValueError("Please enter a valid Salary")
    self.salary = salary
    if('@' not in email):
      raise WrongEmailError("Bro enter Valid Email")
    self.email = email
def show_salary(self):
    print(self.salary)
def show_email(self):
    print(self.email)
```

```python
try:
   p1 = Person(100000,"def#gmail.com")
except WrongEmailError as e:
    print("Wrong Email")
    print("Error message:",e)  
  
```
